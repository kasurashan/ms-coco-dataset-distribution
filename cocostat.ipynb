```python
import pandas as pd
```

```python
annot_area = []
annot_ratio = []
annot_cat = []
for annot in coco.loadAnns(coco.getAnnIds()):
    annot_area.append(annot['area'])
    annot_ratio.append(min(annot['bbox'][3], annot['bbox'][2])/max(annot['bbox'][3], annot['bbox'][2]))
    annot_cat.append(annot['category_id'])
```

```python
import seaborn as sns
```

```python
coco_stat = pd.DataFrame({ 'category':annot_cat, 'ratio':annot_ratio, 'area':annot_area})
coco_stat.head(8)
```

```python
sns.pairplot(coco_stat, diag_kind='kde')
```

```python
coco_stat.describe()
```

```python
print(coco_stat[coco_stat['area']<32**2].describe())
print(coco_stat[(coco_stat['area']<96**2) & (coco_stat['area']>32**2)].describe())
print(coco_stat[coco_stat['area']>96**2].describe())
sns.displot(coco_stat[coco_stat['area']<32**2]['area'], binwidth=100)
sns.displot(coco_stat[(coco_stat['area']<96**2) & (coco_stat['area']>32**2)]['area'], binwidth=100)
sns.displot(coco_stat[coco_stat['area']>96**2]['area'], binwidth=100, kde=True)
sns.displot(coco_stat['area'])
plt.savefig('aaaaaaaf.jpg')
```

```python
fig, ax = plt.subplots(nrows=3)

sns.scatterplot(ax=ax[0], data=coco_stat[coco_stat['area']<32**2], x='ratio', y='area')
sns.scatterplot(ax=ax[1],data=coco_stat[(coco_stat['area']<96**2) & (coco_stat['area']>32**2)], x='ratio', y='area')
sns.scatterplot(ax=ax[2],data=coco_stat[coco_stat['area']>96**2], x='ratio', y='area')
```

```python
sns.boxplot(data=coco_stat[coco_stat['area']>96**2], x='area')
sns.boxplot(data=coco_stat[coco_stat['area']>96**2], x='ratio')
```
